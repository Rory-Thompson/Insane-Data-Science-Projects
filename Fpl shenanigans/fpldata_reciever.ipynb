```python

import requests, json
from pprint import pprint
```

```python
# base url for all FPL API endpoints
base_url = 'https://fantasy.premierleague.com/api/'

# get data from bootstrap-static endpoint
r = requests.get(base_url+'bootstrap-static/').json()

# show the top level fields
pprint(r, indent=2, depth=1, compact=True)
```

```python
players = r['elements']

# show data for first player
pprint(players[0])
```

```python
import pandas as pd
players_df = pd.json_normalize(players)
```

```python
required_columns = ['web_name','assists','goals_scored','minutes','expected_assists','expected_assists_per_90',
                    'expected_goals','expected_goals_per_90','expected_goal_involvements_per_90']
players_df= players_df[required_columns]
```

```python
players_df[players_df['minutes']>500].sort_values('expected_goals_per_90').tail(20)
```
