```python
def get_data(coin_type):
    
    btc_historical_data = cryptocompare.get_historical_price_hour(coin_type, currency='USD', limit=2000)
    df = pd.DataFrame.from_dict(btc_historical_data)
    df['datetime'] = pd.to_datetime(df['time'], unit = 's')
    df['coin'] = coin_type
    return df
##the time should be in hour increments, not dependent on when it was ran. IE alwys 000, 2300 2400 0000 etc
```

```python
coins = ['ETH', 'BTC', 'SOL','ADA', 'BNB','ATOM','PEPE','XRP','AVAX','DOGE','SHIB']
```

```python
df_1 = pd.DataFrame()
for coin in coins:
    df_2 = get_data(coin)
    df_1 = pd.concat([df_1, df_2])
df_1
```

```python
df_1['year'] = df_1['datetime'].dt.year
df_1['month'] = df_1['datetime'].dt.month
df_1['day'] = df_1['datetime'].dt.day
df_1['hour'] = df_1['datetime'].dt.hour
#1024 is the maximum number of partitions.
#we will instead only use day not up to hour,this should still drastically reduce computation time
time_now = datetime.datetime.utcnow().strftime("%d,%m,%Y, %H")
df_1.to_parquet('./cryptodf, '+time_now, partition_cols=['year', 'month', 'day'])
```
